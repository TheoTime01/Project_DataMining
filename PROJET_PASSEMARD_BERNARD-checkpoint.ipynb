```python
from PIL import Image
import numpy
import math
import matplotlib.pyplot as plot
from sklearn import svm
import pandas as pd
import sys
import json
from statistics import mean
%matplotlib inline 

Images=['rouge.jpg', 'vert.jpg', 'bleu.jpg']  #Bibliotheques d'image

num_image=input("quelle image voulez vous ? 1 , 2 ou 3 :")  #choix de l'image à analyser

def CouleurPredominante(img):

    image = Image.open(img)
    plot.imshow(image)
    plot.show()   # affichage de l'image

    numarray = numpy.array(image.getdata(), numpy.uint8) # on récupère la matrice des pixels de l'image

    r=[]  # on initialise les listes de chaque couleur
    v=[]
    b=[]

    for i in range (len(numarray)):
        r.append(numarray[i][0]) #on ajoute a chaque liste de couleurs la couleur qui correspond de chaques pixels
        v.append(numarray[i][1])
        b.append(numarray[i][2])

    moy_r=mean(r) # on fait la moyenne de ces listes pour avoir la couleur moyenne de l'image
    moy_v=mean(v) 
    moy_b=mean(b)

    my_details = {} #initialisation de la liste qui va écrire dans le fichier .json

    print ("moyenne rouge",moy_r)
    print ("moyenne verte",moy_v)
    print ("moyenne bleu",moy_b)

    if moy_r>=moy_v and moy_r>=moy_b: # si La couleur prédominante est le rouge
        print ('La couleur prédominante est le rouge')
        my_details = {'La couleur predominante est le' : 'rouge','valeur moyenne rouge' : str(moy_r),'valeur moyenne vert': str(moy_v),'valeur moyenne bleu': str(moy_b)}

    elif moy_v>moy_r and moy_v>=moy_b:  # si La couleur prédominante est le vert
        print ('La couleur prédominante est le vert')
        my_details = {'La couleur predominante est le' : 'vert','valeur moyenne rouge' : str(moy_r),'valeur moyenne vert': str(moy_v),'valeur moyenne bleu': str(moy_b)}

    elif moy_b>moy_r and moy_b>moy_v:  # si La couleur prédominante est le bleu
        print ('La couleur prédominante est le bleu')
        my_details = {'La couleur predominante est le' : 'bleu','valeur moyenne rouge' : str(moy_r),'valeur moyenne vert': str(moy_v),'valeur moyenne bleu': str(moy_b)}

    with open('Fichier.json', 'w') as json_file:  # on crée un fichier .json
        json.dump(my_details, json_file)   # on écrit dans le fichier 

        
        
    
CouleurPredominante(Images[int(num_image) - 1]) # appele de la fonction avec en paramètre l'image choisi -1 pour etre entre 0 et 2 (3 images)
```
